**Funciones**

```python
def get_my_list(lista_compleja):
  for index in range(0, len(lista_compleja)-1):
    if type(lista_compleja[index]) == list:
      a = lista_compleja[index]
  return(a)

print(get_my_list(lista_compleja))
```

```python
lista_fill = []
for i in range(0,20):
  lista_fill.append(i)
lista_fill
```

```python
index = 0
while index < len(lista_fill):
  print(lista_fill[index])
  index = index + 1 #index += 1
  
```

```python
intervalo = 4
contador = 0
split_list = []
#Para heúristica
#70 - 30, Entrenar y Prrobar
while contador  < len(lista_fill):
  split_list.append(lista_fill[contador:contador+intervalo])
  contador += intervalo
print(split_list)
```

**Diccionarios**

Un diccionario es una estructura de datos en Python que almacena diferentes tipos de datos, y sus valores son obtenidos por llaves (keys). Técnica Hashmap.
Hash function.
Son más rápidos que listas y tuplas.

```python
diccionario = {'nombre': 'Nestor', 'edad':22, 'curso':'Python'}
type(diccionario)
```

```python
diccionario.keys()
```

```python
diccionario.values()
```

```python
for key in diccionario.keys():
  print(diccionario[key])
```
